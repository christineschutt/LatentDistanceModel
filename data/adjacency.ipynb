```python
zeroes = 0
ones = 0
twos = 0
threes = 0
for i in range(len(data)):
    if data.iloc[i,4] == 'not detected':
        zeroes +=1
    elif data.iloc[i,4] == 'rare':
        ones +=1

    elif data.iloc[i,4] == 'uncommon':
        twos+=1

    elif data.iloc[i,4] == 'common':
        threes +=1
```

```python
print(zeroes, ones, twos, threes)
```

## Convert ordinal categories to integers

```python
#ensure that 0 is not used
mapping = {'not detected': 0, 'rare': 1, 'uncommon': 2, 'common': 3}
data['Frequency'] = data['Frequency'].replace(mapping)
data
```

```python
data = data.sort_values(by='Frequency', ascending=False)
data
```

```python
data.iloc[0,4]
```

```python
zeroes = 0
ones = 0
twos = 0
threes = 0
for i in range(len(data)):
    if data.iloc[i,4] == 0:
        zeroes +=1
    elif data.iloc[i,4] == 1:
        ones +=1

    elif data.iloc[i,4] == 2:
        twos+=1

    elif data.iloc[i,4] == 3:
        threes +=1
```

```python
print(zeroes, ones, twos, threes)
```

```python
data['Frequency'].unique()
```

```python
# Group by drug-side effect pairs and count occurrences
adj_matrix = data.pivot_table(index="Stitch flat", columns="Side effect", values="Frequency", aggfunc="max")
adj_matrix = adj_matrix.fillna(0)
adj_matrix.to_csv('adj_matrix_names.csv', sep='\t')
adj_matrix
```
